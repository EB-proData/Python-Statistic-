## 1.3 Décomposition en moyenne mobile et évaluation de la tendance globale 

### 1.3.1 Evaluer la moyenne trimestrielle & représentation graphique avec la tendance

```python
trie_librairie_web["date"] = pd.to_datetime(trie_librairie_web["date"])

# Calculer le chiffre d'affaires par jour
CA_par_jour = trie_librairie_web.resample("D", on="date")["price"].sum()

# Calculer le chiffre d'affaires par semaine
CA_par_semaine = trie_librairie_web.resample("W", on="date")["price"].sum()

# Regrouper les données par trimestre et calculer la moyenne du chiffre d'affaires pour chaque trimestre
CA_par_trimestre = CA_par_semaine.resample("Q").mean()

# Afficher la moyenne du chiffre d'affaires par trimestre
print(CA_par_trimestre)
```

```python
# Effectuer la décomposition de la moyenne mobile
decomposition = seasonal_decompose(CA_par_jour, model="additive")

# Afficher les composantes
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Créer un graphique pour visualiser la décomposition
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(CA_par_jour, label="Chiffre d\'affaires par jour", color="blue")
plt.legend(loc="upper left")

plt.subplot(4, 1, 2)
plt.plot(trend, label="Tendance", color="red")
plt.legend(loc="upper left")

plt.subplot(4, 1, 3)
plt.plot(seasonal, label="Saisonnalité", color="green")
plt.legend(loc="upper left")

plt.subplot(4, 1, 4)
plt.plot(residual, label="Résidus", color="orange")
plt.legend(loc="upper left")
plt.ylim(-4000, 4000)

plt.tight_layout()
plt.show()
```

 <b><span style="text-decoration: underline;">Saisonnalité</span></b> : 

50 : Cela indique une variation saisonnière positive relativement importante. Cela signifie qu'à ce moment-là, le chiffre d'affaires est généralement supérieur à la moyenne en raison de la saisonnalité.

0 : Cela indique qu'il n'y a pas de variation saisonnière à ce moment-là, ou que la variation est négligeable.

-50 : Cela indique une variation saisonnière négative relativement importante. Cela signifie qu'à ce moment-là, le chiffre d'affaires est généralement inférieur à la moyenne en raison de la saisonnalité.

 <b><span style="text-decoration: underline;">Résidus</span></b> :  :

plus proche de 2000 : Cela indique un résidu positif relativement important. Cela signifie qu'à ce moment-là, le chiffre d'affaires est plus élevé que prévu en fonction de la tendance et de la saisonnalité.

0 : Cela indique qu'il n'y a pas d'écart par rapport à la tendance et à la saisonnalité à ce moment-là, ou que l'écart est négligeable.

moins proche de 2000 : Cela indique un résidu négatif relativement important. Cela signifie qu'à ce moment-là, le chiffre d'affaires est plus bas que prévu en fonction de la tendance et de la saisonnalité.

```python
# Décomposition de la moyenne mobile par mois
decomposition = seasonal_decompose(CA_par_mois, model="additive")

# Tendance
trend = decomposition.trend

# Calcul de la variation du CA d'un mois à l'autre
variation_CA = CA_par_mois.diff()

# Identifier les mois avec une hausse de CA
mois_hausse = variation_CA > 0

# Graphique
plt.figure(figsize=(12, 6))

# Chiffre d'affaires moyen par mois
plt.plot(CA_par_mois.index, CA_par_mois, label="Chiffre d'affaires moyen par mois", color="lightblue")

# Tendance
plt.plot(trend.index, trend, label="Tendance", color="blue")

# En rouge les mois avec une hausse de CA
for mois in CA_par_mois.index[mois_hausse]:
    plt.axvline(x=mois, color='r', linestyle='--', alpha=0.5)

# Graphique
plt.title("Décomposition de la moyenne mobile par mois - Tendance")
plt.xlabel("Mois")
plt.ylabel("Chiffre d'affaires moyen par mois")
plt.legend(loc="upper left")
plt.grid(True)
plt.xticks(pd.date_range(start=CA_par_mois.index.min(), end=CA_par_mois.index.max(), freq='MS'), rotation=45)
plt.show()


```

### 1.5.3 Courbe de LORENZ

```python
# Données pour réaliser la courbe Lorenz
sorted_values = total_par_client.sort_values(ascending=True)

# Calcul des pourcentages cumulés des valeurs
cumulative_values = np.cumsum(sorted_values)

# Calcul des pourcentages cumulés des clients
total_sum = cumulative_values.iloc[-1]
cumulative_clients = np.linspace(0, 100, len(cumulative_values))

# Calcul des pourcentages cumulés des valeurs par rapport au total
cumulative_values_percent = cumulative_values / total_sum * 100

# Calcul des pourcentages cumulés des clients par rapport au total
cumulative_clients_percent = cumulative_clients

# Tracer de la courbe de Lorenz
plt.figure(figsize=(8, 6))
plt.plot(cumulative_clients_percent, cumulative_values_percent, marker="o", color="blue")
plt.plot([0, 100], [0, 100], color="red", linestyle="--") 
plt.title("Courbe de Lorenz")
plt.xlabel("Pourcentage cumulé des clients")
plt.ylabel("Pourcentage cumulé du chiffre d'affaires total")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calcul de l'aire sous la courbe de Lorenz
A = np.trapz(cumulative_values_percent / 100, x=cumulative_clients_percent / 100)

# Calcul de l'aire totale sous la ligne de l'égalité parfaite
B = 0.5 

# Calcul du coefficient de Gini
G = (B - A) / B
print("Coefficient de Gini:", G)
```

L'indice de Gini est un indicateur de l'inégalité dans une distribution (ici le chiffre d'affaires). 

Il varie de 0 à 1 où : 

- 0 représente une société parfaitement égalitaire (ici que tous les individus contribuent de maniere égalitaire au chiffre d'affaires) 
- 1 représente une société parfaitement inégalitaire (ici que seule quelques individus contribuent au chiffre d'affaires). 

Ce coefficient se calcule en mesurant l'aire entre la courbe de Lorenz bleue, qui représente la distribution cumulative réelle du chiffre d'affaires, et la courbe de Lorenz rouge, qui représente une répartition parfaitement égalitaire.

Dans notre cas, un coefficient de Gini de 0,4 indique une certaine inégalité dans la distribution du chiffre d'affaires, mais cette inégalité n'est pas extrême car le coefficient est plus proche de 0 que de 1. En d'autres termes, bien que certains individus ou groupes réalisent une part disproportionnée du chiffre d'affaires, la répartition reste relativement équilibrée par rapport à une répartition parfaitement égalitaire.

### 1.5.4 Courte conclusion sur le zoom sur les clients :

La distribution du chiffre d'affaires entre les clients est plus inégalitaire que la répartition égale. 
En d'autres termes, un petit pourcentage de clients contribue à une grande partie du chiffre d'affaires total, ce qui indique une concentration plus élevée du chiffre d'affaires chez certains clients. Cela peut être interprété comme une concentration de la richesse ou du chiffre d'affaires parmi un petit groupe de clients, tandis que la majorité des clients contribuent à une part relativement faible du chiffre d'affaires total.

## 2.1 Lien entre le genre d'un client et les catégories des livres achetés 

Type de données = 2 variables Qualitatives nominales (genre et catégorie)

Nous cherchons donc une **corrélation** entre deux variables de même type

Choix du test porté sur :

Le **test du khi2** pour des variables prenant un nombre fini de valeurs possibles (donc des variables catégorielles). Pour rappel, un test statistique est une méthode permettant d’accepter ou non une hypothèse, appelée hypothèse nulle, selon son adéquation aux données.
À quoi sert le test du khi-2 ?
L’avantage du test khi-deux est sa grande diversité d’utilisation :

- Test d’adéquation à une loi ou une famille de lois définies à priori, par exemple : la taille d’une population suit-elle une loi normale ? :
- Test d’indépendance, exemple : la couleur des cheveux est elle indépendante du sexe ?
- Test d’homogénéité : deux séries de données sont-elles identiquement distribuées?

**Conditions d'utilisation du Test chi²**:

- Taille d'échantillon suffisamment grande : Ici un grand nombre d'observations (687 534 lignes), ce qui est suffisant pour le test du chi-2. 

- Fréquences attendues suffisantes : Les fréquences attendues dans chaque cellule de la table de contingence doivent être supérieures à 5 pour que le test du chi-2 soit valide. 

- Variables catégorielles : Les variables doivent être catégorielles, c'est-à-dire qu'elles doivent être des variables qualitatives et non continues.  

- Indépendance des observations : Chaque observation est indépendante des autres.  


source : https://datascientest.com/test-du-khi-2

```python
# Créer une table de contingence
table_contingence = pd.crosstab(librairie_web_sans_outliers["sex"], librairie_web_sans_outliers["categ"])

# Effectuer le test du chi-2 pour obtenir les fréquences attendues
chi2, p_value, dof, expected = chi2_contingency(table_contingence)

# Vérifier si les fréquences attendues sont supérieures à 5
freq_attendues_superieures_5 = (expected > 5).all()

if freq_attendues_superieures_5:
    print("Les fréquences attendues dans chaque cellule sont supérieures à 5.")
else:
    print("Les fréquences attendues dans au moins une cellule sont inférieures ou égales à 5.")
```

### 2.1.1 Hypotheses d'étude

Hypothèse nulle (H0) : il n'y a pas de corrélation entre le genre et la catégorie de livre achetée. Nos variables sont indépendantes.

Hypothèse alternative (H1) : il y a une corrélation entre le genre et la catégorie de livre achetée. Nos variables sont dépendantes.

### 2.1.2 Test du Khi²

```python
# Effectuer le test du Chi-2
chi2, p_value, dof, expected = chi2_contingency(table_contingence)

alpha = 0.05

# Afficher les résultats
print("Chi² Stat:", chi2)
print("P-value:", p_value)

# Comparer la p-value avec mon seuil alpha
if p_value < alpha:
    print("La p-value est inférieure au seuil alpha de", alpha)
    print("On rejette donc l'hypothèse nulle.")
else:
    print("La p-value est supérieure ou égale au seuil alpha de", alpha)
    print("On ne peut pas rejeter l'hypothèse nulle.")
```

Nous pouvons utiliser le **test paramétrique de l'analyse de variance (ANOVA) ou son pendant non paramétrique - le test de Kruskal Wallis**.

L’analyse de variance (ANOVA) est utilisée pour analyser s’il existe une différence entre les moyennes de plus de deux populations. Ainsi, l’analyse de variance nous permet de déterminer si les moyennes de population de deux groupes ou plus sont différentes en analysant la variabilité entre les moyennes des échantillons.

L’hypothèse nulle de l’analyse de variance est donc que les moyennes de tous les groupes analysés sont égales. Alors que l’hypothèse alternative soutient qu’au moins une des moyennes est différente.

Ainsi, l’analyse de variance est particulièrement utile pour comparer les moyennes de plus de deux groupes, car avec ce type d’analyse, vous pouvez étudier les moyennes de tous les groupes en même temps, au lieu de comparer les moyennes par paires. Nous verrons ci-dessous quels sont les avantages et les inconvénients de l’analyse de variance.

```python
# Test de Kolmogorov-Smirnov pour l'âge
ks_statistic_age, p_value_age = kstest(librairie_web_sans_outliers["age"], 'norm', 
                                       args=(librairie_web_sans_outliers["age"].mean(), 
                                             librairie_web_sans_outliers["age"].std()))

print("Résultats du test de Kolmogorov-Smirnov pour l'âge :")
print("Statistique de test:", ks_statistic_age)
print("p-valeur:", p_value_age)

# Conclusion
if p_value_age < 0.05:
    print("Conclusion: La distribution de ma colonne 'age' n'est pas normale.")
else:
    print("Conclusion: La distribution de 'age' est normale.")
```

```python
# Diviser les données en groupes par catégorie de livre
grouped_data = [librairie_web_sans_outliers[librairie_web_sans_outliers["categ"] == category]["age"].dropna() 
                for category in librairie_web_sans_outliers['categ'].unique()]

# Test de Levene pour l'homogénéité des variances
levene_statistic, p_value_levene = levene(*grouped_data)

print("Résultats du test de Levene pour l'homogénéité des variances:")
print("Statistique de test:", levene_statistic)
print("p-valeur:", p_value_levene)

# Conclusion
if p_value_levene < 0.05:
    print("Conclusion: Il n'y a pas d'homogénéité des variances.")
else:
    print("Conclusion: Il y a homogénéité des variances.")
```
